```python
import pandas as pd
import numpy as np
from ipywidgets import widgets
```

Developing the function to price a zero coupon bond with facevalue 100

```python
def pv(start,end,rate):
    start=pd.to_datetime(start)
    end = pd.to_datetime(end)
    if start>end:
        start=end
    tenor = (end-start).days/365
    return 100/(np.power(1+rate,tenor))
```

Test the function

```python
start=pd.to_datetime('20180101')
end=pd.to_datetime('20200101')
pv(start,end,0.1)
```

```python
title = widgets.Label(value='Zero Coupon Bond Pricer')
maturity_date = widgets.DatePicker(description='Maturity Date',value=pd.to_datetime('2025-01-01'))
issue_date = widgets.DatePicker(description='Issue Date',value=pd.to_datetime('2020-01-01'))
interest_rate = widgets.FloatSlider(min=0,
                                         max=10.0,
                                         step=0.1,
                                         description='Rate(%)')
```

Add Button

```python
button = widgets.Button(description="calculate")
```

Add the interaction: if you click a button => bond is priced=> output is written on screen

```python
def on_button_clicked(b):
    with output:
        output.clear_output()
        print('Bond Price:', np.round(pv(issue_date.value,
                 maturity_date.value,
                 interest_rate.value/100),2))
        
button.on_click(on_button_clicked)
```

Output
