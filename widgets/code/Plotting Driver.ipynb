```python
df = pd.read_csv('s3://lfspf/driver.csv',sep=';')
```

```python
df.head()
```

```python
df.sort_values(by='Time_GPS',ascending=True)
```

```python
set(df.columns)
```

```python
lat_avg = df['Latitude_GPS'].mean()
lon_avg = df['Longitude_GPS'].mean()
```

```python
m = Map(center=(lat_avg, lon_avg), zoom=13)

matter_layer = basemap_to_tiles(basemaps.OpenStreetMap.HOT)
m.add_layer(matter_layer)
m
```

```python
import numpy as np
```

```python
idx = np.arange(0,len(df),100)
```

```python
loc = df.loc[idx,['Latitude_GPS','Longitude_GPS']].values
```

```python
l=[list(i) for i in loc]
```

```python
line = Polyline(
    locations=l,
    color="red" ,
    fill=False
)
```

```python
center = np.mean(np.array(l),axis=0)
```

```python

m = Map(center=(center[0], center[1]), zoom=13)

matter_layer = basemap_to_tiles(basemaps.OpenStreetMap.HOT)
m.add_layer(matter_layer)
m.add_layer(line)
m
```
